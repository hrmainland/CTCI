```python
import heapq
from random import randint, seed
seed(42)
```

```python
rand_list = [randint(0, 200) for _ in range(20)]
print(rand_list)
```

```python
heap = []
for elem in rand_list:
    heapq.heappush(heap, elem)
print(heap)
heapq.nsmallest(3, heap)
```
